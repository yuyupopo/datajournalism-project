```python
import pandas as pd
from pandas import DataFrame

date_d = pd.read_csv('in_out_by_date.csv', encoding='utf-8')
date_dd = DataFrame(date_d)
```

```python
df = pd.DataFrame(date_d.values.repeat(2, axis=0), columns=date_d.columns)
df.columns = ['구분', 'date', 'line', 'code', 'name', 'in_people',
'out_people']
df
```

```python
for i in df.index:
    if i%2 == 0:
        df.loc[i]['in_people'] = df.loc[i]['out_people']
        df.loc[i]['구분'] = '하차'
    else:
        df.loc[i]['구분'] = '승차'
        pass
```

```python
#df = df.drop(['out_people'],axis=1)
df.head(6)
```

```python
with open('date_data.csv','w', encoding='utf-8') as f:
    df.to_csv(f)
```
