```python
import pandas as pd
import numpy as np

data = pd.read_csv("2017_CALL_TAXI2_12MONTH.csv")
```

```python
import numpy as np
import pandas as pd

data2 = {'date':[],'총통화건수':[]}
list_1 = []
list_2 = []

for i in range(20171201, 20171232):
    list_1.append(str(i))
    list_2.append(sum(data.loc[i]['통화건수']))

data2['date'] = list_1
data2['총통화건수'] = list_2
```

```python
data2_df = pd.DataFrame(data2)
data2_df.index = data2['date']
data2_df = data2_df.drop(['date'], axis=1)
list_3 = []

data2_df
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

#import seaborn as sb

figure(num=None, figsize=(15, 10), dpi=80, facecolor='w', edgecolor='k')
plt.plot('date','총통화건수', data=data2, marker = 'o')
plt.xticks(rotation=45)
plt.show()
```
