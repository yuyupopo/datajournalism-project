Naver Open API로는 기간 설정 검색 불가능

네이버 검색 페이지에서 크리스마스 데이트, date_from = 20161201, date_to = 20161225 로 설정한 다음 크롤링, 검색 결과는 최근 1,000개만 제공. 왜 url이 그보다 적게 나왔는지 코드 상으로는 잘 모르겠음...

* date_from: 시작날짜(YYYYMMDD)
* date_to: 마지막날짜(YYYYMMDD)

'크리스마스 데이트' 검색 결과 크롤링

```python
from bs4 import BeautifulSoup
import urllib.request
import json
from collections import OrderedDict

url_list = []
with urllib.request.urlopen("https://search.naver.com/search.naver?where=post&query=%ED%81%AC%EB%A6%AC%EC%8A%A4%EB%A7%88%EC%8A%A4%20%EB%8D%B0%EC%9D%B4%ED%8A%B8&st=sim&sm=tab_opt&date_from=20171201&date_to=20171225&date_option=8&srchby=all&dup_remove=1&post_blogurl=&post_blogurl_without=&nso=so%3Ar%2Ca%3Aall%2Cp%3Afrom20171201to20171225&mson=0") as url:
    doc = url.read()
    soup = BeautifulSoup(doc, 'html.parser')
    divs = soup.find_all("div", class_="thumb thumb-rollover")
    for div in divs:
        url_list.append(div.a["href"])

#1페이지
```

```python
# 2페이지~100페이지
for i in range(1,100):
    with urllib.request.urlopen("https://search.naver.com/search.naver?date_from=20171201&date_option=8&date_to=20171225&dup_remove=1&nso=p%3Afrom20161201to20161225&post_blogurl=&post_blogurl_without=&query=%ED%81%AC%EB%A6%AC%EC%8A%A4%EB%A7%88%EC%8A%A4%20%EB%8D%B0%EC%9D%B4%ED%8A%B8&sm=tab_pge&srchby=all&st=sim&where=post&start={}".format(i*10+1)) as url:
        doc = url.read()
        soup = BeautifulSoup(doc, 'html.parser')
        divs = soup.find_all("div", class_="thumb thumb-rollover")
        for div in divs:
            url_list.append(div.a["href"])

    
len(url_list)
```

```python
url_list
```

```python
with open('171225_search_date.txt', 'w') as f:
    for item in url_list:
        f.write("%s\n" % item)
```
