```python
from bs4 import BeautifulSoup
import urllib.request
from selenium import webdriver
import time


# url = "http://v.media.daum.net/v/20161103113844417"
url = "https://ko.foursquare.com/explore?cat=food&mode=url&near=%EC%84%9C%EC%9A%B8%ED%8A%B9%EB%B3%84%EC%8B%9C%2C%20%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD&nearGeoId=72057594039763784"
#url = "https://ko.foursquare.com/explore?cat=food&mode=url&ne=37.591383%2C127.097912&sw=37.453875%2C126.898098"
driver = webdriver.Chrome(executable_path = r'C:\Users\user\Desktop\chromedriver_win32\chromedriver.exe')
driver.get(url)

found_element = True
while found_element:
    try:
        element = driver.find_element_by_xpath("//button[@class='blueButton']")
        element.click()
        time.sleep(3)

    except Exception:
        found_element = False

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
```

```python
food_link = []

for link in soup.find_all('a'):
    food_link.append(link.get('href'))

len(food_link)
```

```python
food_link[19]
```

```python
food_link_clear = []
for s in food_link:
    if '/v/' in s:
        food_link_clear.append(s)
```

```python
food_link_clear
```

```python
url_foods = []
for i in food_link_clear:
    url_foods.append("https://ko.foursquare.com" + i)
url_foods
```

```python
len(url_foods)
```

```python
for base_url in url_foods:
    with urllib.request.urlopen(base_url) as url:
        doc = url.read()
        soup = BeautifulSoup(doc, "html.parser")
        title = soup.find_all("h1")[0].text
        print(title)
len(title)
```

```python
for base_url in url_foods:
    
    with urllib.request.urlopen(base_url) as url:
        doc = url.read()
        soup = BeautifulSoup(doc, "html.parser")
        location = soup.find_all("span", class_ = "venueNeighborhood")[0].text.strip()
        print(location)
```
